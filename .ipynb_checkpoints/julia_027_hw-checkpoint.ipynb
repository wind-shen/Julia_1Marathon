# Julia 機器學習：GLM 線性迴歸

## 作業 027：波士頓房價預測資料集

請使用 GLM 中的模型，建立一個預測模型來預測波士頓的房價。

```julia
using GLM, RDatasets, MLDataUtils
```

## 讀取資料

```julia
boston = dataset("MASS", "Boston")
first(boston, 10)
```

```julia
indecies = MLDataUtils.shuffleobs(collect(1:nrow(boston)))
train_ind, test_ind = MLDataUtils.splitobs(indecies, at=0.8);
```

```julia
train = boston[train_ind, :]
test = boston[test_ind, :]
```

```julia
show(boston, allcols=true)
```

```julia
ols = GLM.lm(@formula(NOx ~ Crim + Zn + Indus + Chas + LStat + Rm + Age + Dis + Rad + Tax + PTRatio + Black + MedV), train)
```

```julia
predict(ols, test)
```

```julia
GLM.r²(ols)
```

```julia
GLM.adjr²(ols)
```

```julia
using StatsBase, RDatasets
using Gadfly
import Cairo, Fontconfig
```

```julia
 p = plot(boston, x="LStat", y="NOx", xintercept=[mean(boston[!, :LStat])], 
    Geom.vline(color=["blue"]), yintercept=[mean(boston[!, :NOx])], Geom.hline(color=["red"]), Geom.point)
```

```julia
logreg = glm(@formula(NOx ~ Crim + Zn + Indus + Chas + LStat + Rm + Age + Dis + Rad + Tax + PTRatio + Black + MedV), train, Binomial(), LogitLink())
```

```julia
pred = predict(logreg, test)
```

```julia
error(x, y) = ((x > 0.5) ? 1.0 : 0.0) == y
accuracy(xs, ys) = mean(error.(xs, ys))
```

```julia
accuracy(pred, test[!, :NOx])
```
